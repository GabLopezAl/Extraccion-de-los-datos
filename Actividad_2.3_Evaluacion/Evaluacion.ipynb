**Importacion de librerias**

```python
import pandas as pd
import numpy as np
import matplotlib as mt
import seaborn as sns
```

**Identificar codificacion del archivo**

```python
import chardet
with open('Berlin.csv', 'rb') as f:
    result = chardet.detect(f.read())
    print(result['encoding'])
```

**Leer archivo csv**

```python
df = pd.read_csv('Berlin.csv', encoding=result['encoding'])
df
```

```python
# Verificar tipos de datos
df.dtypes
```

**Primer filtro: 5 registros con host_name diferentes**

```python
# Se filtraron los datos por cabecera para mostrar los primeros cinco registros
filtro1 = df.head()
filtro1
```

```python
filtro1.to_csv("Filtro1_5_registros.csv")
```

**Filtro 2: Registros de host que se hayan unido antes del año 2019**

```python
#Se usó la variable host_since para verificar registros antes del año 2019
df["host_since"] = pd.to_datetime(df["host_since"], errors="coerce")
filtro2 = df[df["host_since"] < "2019-01-01"]
filtro2
```

```python
filtro2.to_csv("Filtro2_Registros_Menores_2019.csv")
```

**Filtro 3: Resgitros en donde los host responden a más tardar un día**

```python
# Se usó la variable host_response_time para verificar el tiempo de respuesta del host
filtro3 = df[df["host_response_time"] == "within an hour"]
filtro3
```

```python
filtro3.to_csv("Filtro3_Respuesta_1_dia.csv")
```

**Filtro 4: Resgistros que tengan al menos 5 accommodates**

```python
# Se usó la variable accommodates para verfiifcar registros en dond el valor sea igual a 5
filtro4 = df[df["accommodates"] == 5]
filtro4
```

```python
filtro4.to_csv("Filtro4_Accommodates_5.csv")
```

**Filtro 5: Registros en donde el tipo de cuarto sea igual a Entire home y Private Room**

```python
filtro5 = df[(df["room_type"] == "Entire home/apt") & (df["room_type"] == "Private room")]
filtro5
```

**Filtro 6: Registro que cuneten al menos con 4 camas**

```python
# Se usó la variable beds para filtrar registros que tengan 4 camas
filtro6 = df[df["beds"] == 4]
filtro6
```

```python
filtro6.to_csv("Filtro6_Beds_4.csv")
```

**Filtro 7: Resgitros tengan 1 baño o sean superhost, condicion todos sean instant_bookable**

```python

filtro7 = df[(df["bathrooms"] == 1) | (df["host_is_superhost"] == "t")]
filtro7
```

**Filtro 8: REgistros cuyo review_scores_cleanliness menores a 3 y review_scores_rating mayor a 3**

```python
#Se usó la variable review_scores_cleanliness para mostrar registros en donde el valor sea menor a 3 y
# la variable review_scores_rating para mostrar registros mayores a 3
filtro8 = df[(df["review_scores_cleanliness"] < 3) & (df["review_scores_rating"] > 3)]
filtro8
```

```python
filtro8.to_csv("Filtro8_cleanliness_raiting.csv")
```

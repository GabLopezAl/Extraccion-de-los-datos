**Importación de librerías**

```python
import pandas as pd
import numpy as ny
import matplotlib as mt
import seaborn as sns
```

**Leer dataframe**

```python
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

**Visualizar información del dataframe**

```python
df.info()
```

**Filtro 1: Por Objeto**

```python
# En este script se filtraron los datos usando la variable NOMBRE_VENDEDOR, se usó el nombre de los registros
# que coincidieran con el nombre de LETICIA RAMIREZ HERNANDEZ

filtro1 = df[df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ"]
filtro1
```

```python
# Convertir archivo filtrado a CSV
filtro1.to_csv("Filtro_1_Por_Objeto.csv")
```

**Filtro 2: Por filas**

```python
# En este script se filtaron los datos solo para mostrar los primeros 5 registros por filas

filtro2 = df.iloc[0:5,:]
filtro2
```

```python
# Convertir archivo filtrado a CSV
filtro2.to_csv("Filtro_2_Por_Filas.csv")
```

**Filtro 3: Por columnas**

```python
# En este script se filtraron los datos solo para mostrar las columnas 2,3,4 y 6 respectivamente, las cuales son:
# NOMBRE_VENDEDOR, NOMBRE_CLIENTE, CANT, DESCR

filtro3 = df.iloc[:, [2,3,4,6]]
filtro3
```

```python
# Convertir archivo filtrado a CSV
filtro3.to_csv("Filtro_3_Por_Columna.csv")
```

```python
# Leer dataframe con indice
df1 = pd.read_excel("Detalle precios y productos fabricados 2022.xlsx", index_col = 2)
df1
```

**Filtro 4: Filas vs Columnas**

```python
# En este script se filtraron los datos para conocer el nombre del cliente, la cantidad y el precio unitario de dos
# vendedores en específico: ALFREDO CANELA y AARON NUÑO CORDOBA

filtro4 = df1.loc[["ALFREDO CANELA", "AARON NUÑO CORDOVA"], ["NOMBRE_CLIENTE", "CANT", "PRECIO_UNITARIO"]]
filtro4
```

```python
# Convertir archivo filtrado a CSV
filtro4.to_csv("Filtro_4_Filas_vs_Columnas.csv")
```

**Filtro 5: Por cabecera**

```python
# Este script filtra los datos para mostar los primeros 5 registros que se tienen en el dataframe

filtro5 = df1.head()
filtro5
```

```python
# Convertir archivo filtrado a CSV
filtro5.to_csv("Filtro_5_Por_Cabecera.csv")
```

**Filtro 6: Por Cola**

```python
# Este script filtra los datos para mostrar los 5 últimos registros del dataframe

filtro6 = df1.tail()
filtro6
```

```python
# Convertir archivo filtrado a CSV
filtro6.to_csv("Filtro_6_Por_Cola.csv")
```

**Filtro 7: Por comparación**

```python
# Este script filtra los datos para mostrar los registros en donde los valores de la columna CANT sean mayores a 5

filtro7 = df[df["CANT"] > 5]
filtro7
```

```python
# Convertir archivo filtrado a csv
filtro7.to_csv("Filtro_7_Por_Comparacion.csv")
```

**Filtro 8: Y**

```python
# En este script se filtraron los datos para mostrar solo los registros en donde la variable CANT sea igual a 7
# y la variable PRECIO_UNITARIO sea mayor a 500

filtro8 = df[(df["CANT"] == 7) & (df["PRECIO_UNITARIO"] > 500)]
filtro8
```

```python
# Convertir archivo filtrado a CSV
filtro8.to_csv("Filtro_8_Y.csv")
```

**Filtro 9: O**

```python
# En este script se filtran los datos solo para mostrar aquellos registros en donde el NOMBRE_VENDEDOR sea igual 
# a ALFREDO CANELA ó el MARGEN_UNITARIO_CALCULADO sea mayor a 10000

filtro9 = df[(df["NOMBRE_VENDEDOR"] == "ALFREDO CANELA") | (df["MARGEN_UNITARIO_CALCULADO"] > 10000)]
filtro9
```

```python
# Convertir archvio filtrado a CSV
filtro9.to_csv("Filtro_9_O.csv")
```

**Filtro 10: NOT**

```python
# En este script se filtraron los datos para mostrar todos los registros a excepcion en donde el PRECIO_UNITARIO sea
# menor a 1000 y la CANT menor a 10, estos resgistros con estos datos no se mostrarán

filtro10 = df[~(df["PRECIO_UNITARIO"] < 1000) & ~(df["CANT"] < 10)]
filtro10
```

```python
# Convertir archivo filtrado a CSV
filtro10.to_csv("Filtro_10_NOT.csv")
```
